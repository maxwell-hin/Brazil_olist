```python
state_dict = {
    'Rondônia': 'RO',
    'Acre': 'AC',
    'Amazonas': 'AM',
    'Roraima': 'RR',
    'Pará': 'PA',
    'Amapá': 'AP',
    'Tocantins': 'TO',
    'Maranhão': 'MA',
    'Piauí': 'PI',
    'Ceará': 'CE',
    'Rio Grande do Norte': 'RN',
    'Paraíba': 'PB',
    'Pernambuco': 'PE',
    'Alagoas': 'AL',
    'Sergipe': 'SE',
    'Bahia': 'BA',
    'Minas Gerais': 'MG',
    'Espírito Santo': 'ES',
    'Rio de Janeiro': 'RJ',
    'São Paulo': 'SP',
    'Paraná': 'PR',
    'Santa Catarina': 'SC',
    'Rio Grande do Sul': 'RS',
    'Mato Grosso do Sul': 'MS',
    'Mato Grosso': 'MT',
    'Goiás': 'GO',
    'Distrito Federal': 'DF'
}
```

```python
swapped_dict = {v: k for k, v in state_dict.items()}

def change_full(state):
    return swapped_dict[state]

customers['customer_state_full'] = customers['customer_state'].apply(change_full)
```

```python
customers.to_csv('customers_with_full.csv', index=False)
```

```python
def top_product(df):
    top_list = ['health_beauty', 'bed_bath_table', 'sports_leisure', 'computers_accessories','watches_gifts']
    if df['product_category_name_english'] in top_list:
        return df['product_category_name_english']
    else:
        return 'Others'
    
product_cat['product_cat'] = product_cat.apply(top_product, axis=1)
    
```

```python
product_cat.to_csv('product_cat.csv', index=False)
```
